```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("data.csv")

time = np.array(df["Time (s)"])
V = np.array(df["Voltage (V)"])
R = np.array(df["Red"])
G = np.array(df["Green"])
B = np.array(df["Blue"])

df.head()
```

```python
TrialName = "Testing 119"
```

# Graph All Data

```python
fig, axs = plt.subplots(3, 1, sharex = 'col', figsize = (10,10))

axs[0].plot(time, R, color='red', label='Red')
axs[0].plot(time, G, color='green', label = 'Green')
axs[0].plot(time, B, color='blue', label = 'Blue')
axs[0].set_xlim(0)
axs[0].set_ylim(0,180)
axs[0].set_xlabel("Time (s)")
axs[0].set_ylabel("RGB Magnitude")
#axs[0].set_title(f"RGB Results [{TrialName}]")
axs[0].legend()

axs[1].plot(time, V, label = 'Voltage', color='purple')
axs[1].set_xlabel("Time (s)")
axs[1].set_ylabel("Voltage (V)")
#axs[1].set_title(f"Voltage Results [{TrialName}]")
axs[1].legend()

ax1 = axs[2]
ax2 = ax1.twinx()

# Copy RGB lines from subplot 0
for line in axs[0].lines:
    ax1.plot(
        line.get_xdata(),
        line.get_ydata(),
        color=line.get_color(),
        label=line.get_label()
    )

ax1.set_ylabel('RGB Magnitude [Max 180]')
ax1.set_ylim(0, 180)

# Copy Voltage line from subplot 1
for line in axs[1].lines:
    ax2.plot(
        line.get_xdata(),
        line.get_ydata(),
        color=line.get_color(),
        label=line.get_label()
    )

ax2.set_ylabel('Voltage (V)')
ax2.set_xlabel('Time (s)')

# Combined legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
```

# Voltage and RGB Same Plot (Save plot as PNG)

```python
fig, ax1 = plt.subplots()

# Plot RGB data
ax1.plot(time, R, color='red', label='Red')
ax1.plot(time, G, color='green', label = 'Green')
ax1.plot(time, B, color='blue', label = 'Blue')
ax1.set_ylabel('RGB Magnitude [Max 180]')
ax1.set_xlabel("Time (s)")
ax1.set_ylim(0, 180)


# Plot voltage data
ax2 = ax1.twinx()
ax2.plot(time, V, color='purple', label='Voltage')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Voltage (V)')

#title and saving
Title = TrialName
plt.title(Title)
plt.savefig(f"./Plots/{Title}")

```
